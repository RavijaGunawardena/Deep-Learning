```python
train_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train"
test_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/test"
val_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/val"

print("Train set:\n========================================")
num_pneumonia = len(os.listdir(os.path.join(train_dir, 'PNEUMONIA')))
num_normal = len(os.listdir(os.path.join(train_dir, 'NORMAL')))
print(f"PNEUMONIA={num_pneumonia}")
print(f"NORMAL={num_normal}")

print("Test set:\n========================================")
print(f"PNEUMONIA={len(os.listdir(os.path.join(test_dir, 'PNEUMONIA')))}")
print(f"NORMAL={len(os.listdir(os.path.join(test_dir, 'NORMAL')))}")

print("Validation set:\n========================================")
print(f"PNEUMONIA={len(os.listdir(os.path.join(val_dir, 'PNEUMONIA')))}")
print(f"NORMAL={len(os.listdir(os.path.join(val_dir, 'NORMAL')))}")

pneumonia = os.listdir("../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/PNEUMONIA")
pneumonia_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/PNEUMONIA"

plt.figure(figsize=(20, 15))

for i in range(9):
    plt.subplot(3, 3, i + 1)
    img = plt.imread(os.path.join(pneumonia_dir, pneumonia[i]))
    plt.imshow(img, cmap='gray')
    plt.axis('off')
#     print (os.path.join(pneumonia_dir, pneumonia[i]))
plt.tight_layout()
```

```python
normal = os.listdir("../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/NORMAL")
normal_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/NORMAL"

plt.figure(figsize=(20, 15))

for i in range(9):
    plt.subplot(3, 3, i + 1)
    img = plt.imread(os.path.join(normal_dir, normal[i]))
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    
plt.tight_layout()
```

```python
normal_img = os.listdir("../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/NORMAL")[0]
normal_dir = "../FYP_Deep Learning_SET/Pneumonia Dataset/archive/chest_xray/train/NORMAL"
sample_img = plt.imread(os.path.join(normal_dir, normal_img))
plt.imshow(sample_img, cmap='gray')
plt.colorbar()
plt.title('Raw Chest X Ray Image')

print(f"The dimensions of the image are {sample_img.shape[0]} pixels width and {sample_img.shape[1]} pixels height, one single color channel.")
print(f"The maximum pixel value is {sample_img.max():.4f} and the minimum is {sample_img.min():.4f}")
print(f"The mean value of the pixels is {sample_img.mean():.4f} and the standard deviation is {sample_img.std():.4f}")
```

```python
sns.displot(sample_img.ravel(),
            label=f"Pixel Mean {np.mean(sample_img):.4f} & Standard Deviation {np.std(sample_img):.4f}", kde=False, height=4, aspect=2)
plt.legend(loc='upper center')
plt.title('Distribution of Pixel Intensities in the Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('# Pixels in Image')
```
